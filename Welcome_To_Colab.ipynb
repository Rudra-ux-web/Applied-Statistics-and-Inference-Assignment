**1.What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making**?

Ans.
Type I and Type II Errors in Hypothesis Testing and Their Impact on Decision-Making
In hypothesis testing, we make decisions about a population based on sample data. We begin with two assumptions: the null hypothesis (H₀), which states that there is no effect or no difference, and the alternative hypothesis (H₁), which states that there is an effect or a difference. Because decisions are based on limited sample information, mistakes may occur. These mistakes are known as Type I and Type II errors.
Type I Error (α Error)
A Type I error happens when the null hypothesis is true, but we reject it by mistake. It is also known as a false positive.
Example: Concluding that a medicine works when actually it does not.
The probability of making a Type I error is called α (alpha), also known as the significance level, commonly set at 0.05 or 0.01. A Type I error can lead to accepting wrong ideas, using ineffective treatments, or implementing policies based on incorrect conclusions.
Type II Error (β Error)
A Type II error occurs when the null hypothesis is false, but we fail to reject it. This is a false negative.
Example: Failing to detect that a medicine works when it actually does.
The probability of this error is called β (beta), and 1 – β is the power of the test, meaning the ability to detect a true effect. A Type II error may cause us to miss useful discoveries or allow defective products to pass quality tests.
Relationship Between Type I and Type II Errors
These two errors are related. Reducing the chance of a Type I error (by choosing a very small α) usually increases the chance of a Type II error, and vice-versa. To reduce both errors, researchers often use larger sample sizes and better experimental designs.
Impact on Decision-Making
Both errors affect real-life decisions in fields like medicine, business, science, and public policy.
•	Impact of Type I Error: Leads to false alarms. We may accept wrong theories, approve harmful medicines, or waste resources on things that do not work.
•	Impact of Type II Error: Leads to missed opportunities. We may ignore useful treatments, fail to detect problems, or miss real effects that could improve decision-making.

**2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?**


Ans.
P-value in Hypothesis Testing and Its Interpretation
In hypothesis testing, the P-value is an important statistical measure that helps us decide whether to reject the null hypothesis (H₀). The null hypothesis generally assumes that there is no effect, no difference, or no relationship in the population. The P-value tells us how likely it is to observe the sample results, or something more extreme, if the null hypothesis is actually true.
Definition of P-value
The P-value is the probability of obtaining the observed results (or more extreme results) assuming that the null hypothesis is true. It is calculated from the sample data and is used to judge the strength of evidence against the null hypothesis.
How to Interpret the P-value
The interpretation of the P-value is based on how small or large it is:
1.	If P-value is small (usually ≤ 0.05):
o	There is strong evidence against the null hypothesis.
o	The observed data is very unlikely to occur if H₀ were true.
o	We reject the null hypothesis.
o	This suggests that a significant effect or difference exists.
2.	If P-value is large (greater than 0.05):
o	There is weak evidence against the null hypothesis.
o	The data could easily occur if H₀ were true.
o	We fail to reject the null hypothesis.
o	This means we do not have enough evidence to claim a significant effect or difference.
Example of Interpretation
Suppose we get a P-value of 0.03. This means there is a 3% chance of getting the sample result assuming that the null hypothesis is true. Since 0.03 is less than 0.05, we reject H₀.
If the P-value is 0.20, it means the sample results could occur 20% of the time if H₀ is true, so we do not reject H₀.
Important Points About P-value
•	A low P-value does not prove that the alternative hypothesis is true; it only shows that the sample data is unlikely under H₀.
•	A high P-value does not prove that the null hypothesis is true; it only means there is not enough evidence to reject it.
•	The P-value depends on sample size: larger samples often lead to smaller P-values.


**3. Explain the difference between a Z-test and a T-test, including when to use each**


Ans.
Difference Between Z-test and T-test and When to Use Each
In statistics, both the Z-test and the T-test are used to test hypotheses about population means. Although they are similar in purpose, they differ in the type of data required and the conditions under which they are applied. Understanding the differences helps in choosing the appropriate test for correct decision-making.
1. Z-test
A Z-test is a statistical test used when the sample size is large and the population standard deviation is known.
Conditions for Using Z-test:
1.	Large sample size (n ≥ 30).
2.	Population standard deviation (σ) is known.
3.	The data should be approximately normally distributed or the sample should be large enough for the Central Limit Theorem to apply.
Purpose of Z-test:
•	To compare a sample mean with a population mean
•	To compare two population means (for large samples)
•	To test proportions

Example:
If we know the population standard deviation and we take a sample of 100 students to test whether their average marks differ from the national average, we use a Z-test.
2. T-test
A T-test is used when the sample size is small and the population standard deviation is not known. In this case, we use the sample standard deviation as an estimate.
Conditions for Using T-test:
1.	Small sample size (n < 30).
2.	Population standard deviation is unknown.
3.	The data must follow a normal distribution or close to normal.
Types of T-tests:
•	One-sample T-test
•	Two-sample (independent) T-test
•	Paired sample T-test
Example:
If we collect a small sample of 15 employees to test whether their average salary differs from an assumed population value, we use a T-test because the population standard deviation is not known.



4.	**What is a confidence interval, and how does the margin of error influence its width and interpretation?**

Ans.
Confidence Interval and the Role of Margin of Error
A confidence interval is a range of values used to estimate an unknown population parameter, such as a population mean or proportion. Instead of giving just one value (called a point estimate), a confidence interval gives a range where we believe the true value lies, based on sample data. It also comes with a confidence level, like 90%, 95%, or 99%, which tells us how sure we are that the interval contains the true value.
Example of a Confidence Interval
If we say that the average height of students is 160 cm to 170 cm at 95% confidence, it means we are 95% sure that the true average height of the population lies within this range.
Margin of Error
The margin of error (ME) is the amount added and subtracted from the sample estimate to form the confidence interval.
It is calculated using:
•	the standard error
•	the critical value (Z or t value)
•	the sample size
The formula for a confidence interval is usually:
Estimate ± Margin of Error
How Margin of Error Affects the Width
The margin of error directly controls how wide or narrow the confidence interval is:
1.	Large Margin of Error → Wide Confidence Interval
o	A wide interval means less precision.
o	We are less certain about the exact value, but more confident that the interval covers the true value.
2.	Small Margin of Error → Narrow Confidence Interval
o	A narrow interval means higher precision.
o	We are more specific in our estimate, but only if the sample supports it.
Factors That Influence Margin of Error
1.	Sample size:
o	Larger sample → smaller margin of error
o	Smaller sample → larger margin of error
2.	Variability in data:
o	More variation → larger margin of error
3.	Confidence level:
o	Higher confidence level (like 99%) → larger margin of error
o	Lower confidence level (like 90%) → smaller margin of error


**5.Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?**



Ans.
Purpose and Assumptions of ANOVA, and How It Extends Hypothesis Testing
ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups at the same time. Instead of doing multiple t-tests for each pair of groups, ANOVA allows us to check whether there is any significant difference among all the group means in one single test. This saves time and also reduces the chances of making errors.
Purpose of ANOVA
1.	To compare more than two group means
ANOVA tells us whether at least one group mean is different from the others.
2.	To analyze the effect of one or more factors
For example, comparing the average marks of students from three different classes.
3.	To avoid multiple t-tests
If we use multiple t-tests for many groups, the error rate increases. ANOVA controls this by using one overall test.
4.	To check variation within and between groups
It compares:
o	Variation between the groups
o	Variation within each group
If the between-group variation is much larger, it suggests the means are not equal.
Assumptions of ANOVA
ANOVA works properly only when certain conditions (assumptions) are met:
1.	Independence of observations
Each sample or group must be selected independently.
2.	Normality
The data in each group should be approximately normally distributed.
3.	Homogeneity of variances (Equal variances)
All groups should have roughly the same variance.
This is also called the assumption of homoscedasticity.
If these assumptions are violated, the results of ANOVA may not be reliable.
How ANOVA Extends Hypothesis Testing to More Than Two Groups
In simple hypothesis testing (like a t-test), we normally compare only two groups. But when there are three, four, or more groups, using separate t-tests becomes complicated and increases the chance of Type I error.
ANOVA solves this by:
1.	Using one null hypothesis:
o	H₀: All group means are equal
o	H₁: At least one group mean is different
2.	Using the F-statistic
ANOVA calculates an F-value by comparing the variance between groups to the variance within groups.
3.	Providing one overall test for all groups
If the F-test is significant (p-value < 0.05), we reject the null hypothesis and conclude that at least one group differs.
4.	Allowing follow-up tests
If the ANOVA result is significant, we can perform post-hoc tests (like Tukey test) to find exactly which groups differ.









```
##6.Write a Python program to perform a one-sample Z-test and interpret the
#result for a given dataset.

import numpy as np
from scipy.stats import norm

# Sample dataset
data = [52, 55, 49, 57, 60, 53, 56, 58, 54, 59]

# Given population mean
population_mean = 50

# Population standard deviation (assumed known)
population_sd = 5

# Calculate sample mean
sample_mean = np.mean(data)

# Sample size
n = len(data)

# Calculate Z-statistic
z_value = (sample_mean - population_mean) / (population_sd / np.sqrt(n))

# Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_value)))

print("Sample Mean:", sample_mean)
print("Z-value:", z_value)
print("P-value:", p_value)
```

```
# 8.Generate multiple samples from a non-normal distribution and implement
#the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate data from a non-normal distribution (Exponential)
population = np.random.exponential(scale=1, size=100000)

# Step 2: Draw multiple samples and compute sample means
sample_means = []
sample_size = 50  # size of each sample
num_samples = 1000  # number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot the original population distribution (non-normal)
plt.hist(population, bins=50, alpha=0.7, edgecolor='black')
plt.title("Original Distribution (Exponential - Non Normal)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 4: Plot the distribution of sample means (CLT)
plt.hist(sample_means, bins=30, alpha=0.7, edgecolor='black')
plt.title("Distribution of Sample Means (CLT Applied)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()


```

```
## 9.Write a Python function to calculate and visualize the confidence interval
##for a sample mean.


import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def confidence_interval_visualization(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for a sample mean.

    Parameters:
    - data: list or numpy array of sample data
    - confidence: confidence level (default is 0.95 for 95% CI)
    """
    # Convert data to numpy array
    data = np.array(data)

    # Sample statistics
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean

    # Calculate the confidence interval
    ci_range = sem * stats.t.ppf((1 + confidence) / 2, df=n-1)
    lower = mean - ci_range
    upper = mean + ci_range

    print(f"Sample mean: {mean:.3f}")
    print(f"{int(confidence*100)}% confidence interval: ({lower:.3f}, {upper:.3f})")

    # Visualization
    plt.figure(figsize=(8, 5))
    plt.errorbar(x=0, y=mean, yerr=ci_range, fmt='o', color='blue', capsize=10, label='Confidence Interval')
    plt.scatter(0, mean, color='red', label='Sample Mean', zorder=5)
    plt.xlim(-1, 1)
    plt.xticks([])
    plt.ylabel('Value')
    plt.title(f'{int(confidence*100)}% Confidence Interval for Sample Mean')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
sample_data = [12, 15, 14, 10, 13, 16, 11, 14, 15, 13]
confidence_interval_visualization(sample_data, confidence=0.95)
```

```
## 10.Perform a Chi-square goodness-of-fit test using Python to compare
##observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example data)
observed = np.array([50, 30, 20])  # Counts observed in each category

# Expected frequencies (based on theory or assumption)
expected = np.array([40, 40, 20])  # Expected counts in each category

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Display results
print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")

```
